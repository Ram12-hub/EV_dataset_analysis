```python
## Task 4: Build an EV recommendation class. The class should allow users to input their budget,desired range, and battery capacity. The class 
## should then return the top three EVEVs matching their criteria. (8+8 Marks)

class EVRecommendation:
    def __init__(self, dataset):
        self.df = dataset
    def recommend(self, user_budget, min_range, min_batteryCapacity):
        filter_df = self.df[ (self.df['Minimal price (gross) [PLN]'] <= user_budget) & (self.df['Range (WLTP) [km]']>= min_range) &
        (self.df['Battery capacity [kWh]'] >= min_batteryCapacity) ]

        if filter_df.empty:
            return "Vehicles are not available with your mentioned criterias please try to another...!"
        
        sorted_df = filter_df.sort_values(by=['Range (WLTP) [km]', 'Battery capacity [kWh]'], ascending = [False, False])

        Top3_evs = sorted_df.head(3)

        return Top3_evs[['Car full name', 'Make', 'Model','Minimal price (gross) [PLN]', 'Battery capacity [kWh]', 'Range (WLTP) [km]']].reset_index(drop = True)
        


EV_recome = EVRecommendation(df)

user_budget = int(input("Pls Enter your budget..."))
user_desired_range = int(input("Pls Eneter your desired range...."))
user_req_battery_cap = int(input("Pls Enter required Battery capacity you want..."))

Recommended_Evs = EV_recome.recommend(user_budget, user_desired_range,user_req_battery_cap )
Recommended_Evs

## Desciption :
## Here class was built based on their budget, range and battery capacity in this case customers can choose the vehicles with their hand money without 
## any confusions simply inoput their budget and desired range and then battery capacity they desire after that top 3 will listed they can choose
## their vehicles here ntg more to mention.
```

```python
## Task 5: Inferential Statistics – Hypothesis Testing: Test whether there is a significant difference in the average Engine power [KM] of vehicles 
## manufactured by two leading manufacturers i.e Tesla and Audi. What insights can you draw from the test results? Recommendations and 
## Conclusion: Provide actionable insights based on your analysis. (Conduct a two sample t-test using ttest_ind from scipy.stats module) (16 Marks)


from scipy.stats import ttest_ind

Tesla_Engine_power = df[df["Make"] == "Tesla"]['Engine power [KM]']
Audi_Engine_power = df[df["Make"] == "Audi"]["Engine power [KM]"]

t_test, p_value = ttest_ind(Tesla_Engine_power, Audi_Engine_power, equal_var = False)

print(f"T_stats : {t_test}")
print(f"P_stats : {p_value}")

alpha = 0.05

if p_value<alpha:
    print("Reject null hypothesis : There is a significant difference in avg engine power of Audi and Tesla")
else:
    print("We fail to reject null hypothesis : There is no significant difference in avg engine power of Audi and Tesla")

# INSIGHTS  FROM THE RESULTS

# 1) Here we can see the T-stats values 1.973 and the p-value is 0.107 it is greater than the significance level of value Alpha=0.05
# 2) This means there is no evidence to say that there is a signigicant difference in Tesla and Audi of AVG engine powers.
# 3) so the p-value is greater than 0.05,so that we can say, we fail to reject the null hypothesis, which means the 
#    average engine powers of Tesla and Audi vehicles are equal.
# 4) Engine power is falied to differentiate the Audi and tesla vehicles so we can check with other aspects.

# RECOMMENDATIONS 

# SO clearly showing there is no significant difference in b/w Audi and Tesla of avg engine powers, so focus on other  features between Tesla and Audi for example range and drive types 
#  and battery capacities like this there are lot of features so we go through thoes type of elements as well.
# Design and Features: Promoting unique aspects like interior design, autonomous driving capabilities, or charging network and infotaiment features such as 
# more they can introduce.
# They can try with other performance factors, such as acceleration , top speed, or energy efficiency.
# they can try to compare with the avg price of bothe the vehicles as well to differentiate
# And they explore the engine power how relation with the prices of the vehicles and then analyse some patterns that may can influence the customer preferences.

# CONCLUSION

# " We got clear evidence that there is no significant difference in the average engine power of Tesla and Audi vehicles. so they can avoid engine power 
#   to differentiate bothe the vehicles and there are lot of other features as suggested above like in between prices and range and battery capacities 
#   and other infitainment featues they can use to differentiate and reach out to the customers.
# "

## Description 
#  " here null hypothesis is defined as there is no difference in the mean engine power of Tesla and Audi and alternative hypothesis defiend as quite opposite to
#    null hypothesis that is significant difference is there between Audi and Tesla mean engine powers as per the stats we doesn't have enough evidance to reject 
#    null hypothesis that means there is no significant difference in the mean engine powers of Audi and Tesla and then nothing more to mention here points
#    are covered under insights and recommendations that's more than enough
# "
```
